```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = fetch_openml('car', as_frame=True).frame
```

```python
data
```

```python
columms_to_encode = ['lug_boot', 'safety']

encoder = OrdinalEncoder(
    categories=[['big', 'med', 'small'], ['high', 'med', 'low']]
)

data[columms_to_encode] = encoder.fit_transform(data[columms_to_encode])
```

```python
data
```

```python
encoder.inverse_transform(data[columms_to_encode])
```

```python
data_2 = fetch_openml('adult' ,as_frame=True).frame
data_2
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data_2.occupation.value_counts()
```

```python
import pandas as pd

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

encoded_values = encoder.fit_transform(data_2[['occupation', 'race']])
new_cols =  encoder.get_feature_names_out(['occupation','race'])
```

```python
new_cols
```

```python
df_encoded = pd.DataFrame(encoded_values, columns=new_cols, index=data_2.index)
data_final = pd.concat(
    [data_2.drop(columns=['occupation' , 'race']), df_encoded]  , axis=1 
)
```
