```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
X, _ = make_blobs(n_samples=5000, centers=5, random_state=10)

X
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=10)
kmeans.fit(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c= kmeans.labels_)
```

```python
from sklearn.datasets import make_moons
X, _ = make_moons(n_samples=5000, noise=0.08, random_state=10)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.2)
dbscan.fit(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan.labels_)
```
